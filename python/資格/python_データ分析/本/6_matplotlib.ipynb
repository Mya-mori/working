```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [1, 4, 9]
y2 = [3, 1, 2]
labels = ["a", "b", "c"]

width = 0.4
ax.bar(x, y1, width = width, tick_label = labels, label = "y1")

x2 = [num + width for num in x]
ax.bar(x2, y2, width = width, tick_label = labels, label = "y2")

ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [1, 4, 9]
y2 = [3, 1, 2]
labels = ["a", "b", "c"]

y_total = [num1 + num2 for num1, num2 in zip(y1, y2)]

ax.bar(x, y_total, tick_label = labels, label = "y1")
ax.bar(x, y2, label = "y2")
ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

np.random.seed(123)
x = np.random.rand(100)
y = np.random.rand(100)

ax.scatter(x[0:20], y[0:20], marker = "v", label = "training down")
ax.scatter(x[20:40], y[20:40], marker = "^", label = "training up")
ax.scatter(x[40:60], y[40:60], marker = "s", label = "square")
ax.scatter(x[60:80], y[60:80], marker = "o", label = "circle")
ax.scatter(x[80:100], y[80:100], marker = "x", label = "x")

ax.legend()

plt.show()
```

```python

np.random.seed(123)
mu = 100 # 平均
sigma = 15 # 標準偏差
x = np.random.normal(mu, sigma, 10000000)

fig, ax = plt.subplots()

n, bins, patches = ax.hist(x)

plt.show()
```

```python
for i, num in enumerate(n):
    print("{:.2f} - {:.2f}".format(bins[i], bins[i+1], num))
```

```python
np.random.seed(123)
x = np.random.randn(10000)

fig, ax = plt.subplots()

counts, edges, patches = ax.hist(x, bins = 25)

x_fit = (edges[:-1] + edges[1:]) / 2

y = 10000 * np.diff(edges) * np.exp(-x_fit**2 / 2) / np.sqrt(2 * np.pi)

ax.plot(x_fit, y)

plt.show()
```
